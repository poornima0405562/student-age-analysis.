# Student Age Analysis

## Objective
To analyze student age data using Python,Pandas & basic statistics.

## Dataset
A small dataset containing student names and ages.

```python
import pandas as pd
data={   
    "Name":['poornima','chathurika','ruth','chanika'],
    "Age":[24,23,22,24]
}
df=pd.DataFrame(data)  
df                                                                                                                                                                                                                                                                                                   
```

```python
df['Age'].mean()
df['Age'].median()
df['Age'].std()
```

```python
import sqlite3

conn = sqlite3.connect("students.db")
df.to_sql("students",conn,if_exists="replace",index=False)

pd.read_sql("SELECT*FROM students",conn)
```
